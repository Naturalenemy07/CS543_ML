```python
import pandas as pd
```

```python
tf = pd.read_csv('dataset_train.csv')
```

```python
tf.head(11)
```

```python
failures = pd.DataFrame([['Air leak','2022-01-01 06:00:00', '2022-01-01 06:00:04'],['Oil leak', '2022-01-01 06:00:07', '2022-01-01 06:00:09']], columns=['Type', 'Start','End'])
```

```python
failures
```

```python
tf_fails = pd.DataFrame(tf['timestamp'].copy(deep=True))
```

```python
tf_fails
```

```python
tf_fails['Failed'] = 0
```

```python
tf_fails
```

```python
end_index = tf_fails[tf_fails['timestamp'] == failures['End'][0]].index[0] + 1
start_index = tf_fails[tf_fails['timestamp'] == failures['Start'][0]].index[0]

air_leak_indices = list(range(start_index, end_index))

tf_fails['Failed'][air_leak_indices] = 1
```

```python
tf_fails.head(11)
```

```python
tf_fails.Failed = 0
```

```python
tf_fails
```

```python
for i in range(failures.shape[0]):
    end_index = tf_fails[tf_fails['timestamp'] == failures['End'][i]].index[0] + 1
    start_index = tf_fails[tf_fails['timestamp'] == failures['Start'][i]].index[0]
    tf_fails['Failed'][start_index:end_index] = 1
```
